```python
fixtures_df = catalog.load("fixtures_primary")
fixtures_df.head()
```

```python
clubs_teams_df = catalog.load("clubs_teams_primary")
clubs_teams_df.head()
```

```python
c = ["home_club_id", "division", "time"]
s = fixtures_df["time"].notna()
fixtures_time = fixtures_df.loc[s, c].drop_duplicates()
fixtures_time = fixtures_time.rename(columns = {"home_club_id":"club_id", "time":"game_time"})
```

```python
c = ["club_id", "division", "game_time"]
clubs_teams_time = clubs_teams_df[c]
```

```python
times = clubs_teams_time.merge(fixtures_time, on=["club_id", "division"], suffixes=("_ct", "_f"))
```

```python
diff = ~(times["game_time_ct"] == times["game_time_f"])
```

```python
scope = ~fixtures_df["played"]

r = fixtures_df.loc[scope].groupby(["home_club_id", "division"])["time"].apply(lambda x: len(x.unique())).sort_values()
r.loc[r>1]
```

```python
scope = ~fixtures_df["played"]
game_time = fixtures_df.loc[scope].groupby(["home_club_id", "division"], as_index=False)["time"].apply(lambda x: x.mode())
game_time = game_time.rename(columns={0:"time_most_frequent"})
```

```python
fixtures_df = fixtures_df.merge(game_time, on=("home_club_id","division"))
```

```python
fixtures_df["time"] = fixtures_df[['time', 'time_most_frequent']].bfill(axis=1).iloc[:, 0]
```
